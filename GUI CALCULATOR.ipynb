```python
import tkinter as tk  #import tkinter module
import math #importing math to evaluate mathematical functions.

def hcf_gcd(a,b): # function to calculate gcd of two numbers
    return math.gcd(a,b)

def lowest_multiple(a,b): # function to calculate lcm of two numbers
    return math.lcm(a,b)

def expo(a,b): # function to calculate exponent of two numbers
    return math.pow(a,b)

def logarithmic(a,b): # function to calculate logarithmic of a number
    return math.log(a,b)


def calculate(): #function to define the calculate operation
    first_number = int(first_number_entry.get()) #getting the values entered by user
    second_number = int(second_number_entry.get())
    operator = operator_entry.get()

    if operator == "1": #perform the calculation based on user's choice
        result = first_number + second_number
        result_label.config(text="Result : " + str(result))
    elif operator == "2":
        result = first_number - second_number
        result_label.config(text="Result : " + str(result))
    elif operator == "3":
        result = first_number * second_number
        result_label.config(text="Result : " + str(result))
    elif operator == "4":
        result = first_number / second_number
        result_label.config(text="Result : " + str(result))
    elif operator == "5":
        result = hcf_gcd(first_number, second_number)
        result_label.config(text="Result : " + str(result))
    elif operator == "6":
        result = lowest_multiple(first_number, second_number)
        result_label.config(text="Result : " + str(result))
    elif operator == "7":
        result = expo(first_number,second_number)
        result_label.config(text="Result : " + str(result))
    elif operator == "8":
        result = logarithmic(first_number, second_number)
        result_label.config(text="Result : " + str(result))
    else:
        result_label.config(text="Invalid operator")

# GUI setup
root = tk.Tk() #create a GUI window
root.configure(background="light green") #setting the background color of GUI window
root.title("Calculator") #setting the title of GUI window
root.geometry("460x400") #setting the configuration of GUI window

#creating the labels and entry fields for the user to enter the number and operators
first_number_label = tk.Label(root, text="First Number")
first_number_label .pack()
first_number_entry = tk.Entry(root)
first_number_entry.pack()

second_number_label = tk.Label(root, text="Second Number")
second_number_label .pack()
second_number_entry = tk.Entry(root)
second_number_entry.pack()

operator_label = tk.Label(root, text="Operations \n ---------- \n  1-Addition \n 2-Subtraction \n 3-Multiplication \n 4-Division \n 5-HCF \n 6-LCM \n 7-Exponent \n 8-Log ")
operator_label .pack()
operator_entry = tk.Entry(root)
operator_entry.pack()

calculate_button = tk.Button(root, text="Calculate", command=calculate) #creating the calculate button
calculate_button.configure(background="pink") #setting the color of calculate button
calculate_button.pack()

result_label = tk.Label(root, text="Result") #label to display the final result
result_label.pack()

root.mainloop() #execute the main window
```
